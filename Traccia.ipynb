Un negozio di animali domestici vuole fare uno sconto 
ai clienti che acquistano un animale (o più) e almeno

mentre gli animali sono esclusi.

Scrivete la funzione `discount(prices, isPet, nItems)`
che calcoli lo sconto sulla base delle informazioni
ricevute sulla vendita in esame, costituita da `nItems`
articoli: per l'articolo i\-esimo, `prices[i]` è 
il prezzo prima dell'eventuale sconto e `isPet[i]` è *True*
se l'articolo i\-esimo è un animale.

Scrivete un programma che chieda al cassiere di digitare 
tutti i prezzi, ciascuno seguito da una *Y* se si tratta 
di un animale e da una *N* per tutti gli altri articoli,
usando il prezzo *\-1* come sentinella. Memorizzate in una
lista i dati acquisiti, poi invocate la funzione che avete
progettato e visualizzate lo sconto.

```python
def dirOnly(istanza):
    metodi = dir(istanza)
    risultato = []
    for m in metodi:
        if not m.startswith('_'):
            risultato.append(m)
    return(risultato)        
```

```python
def discount(prices, isPet):
    pets = sum(isPet)
    altri = sum(prices[i] for i in range(len(prices)) if not isPet[i])
    n_altri = len(prices) - pets

    if pets >= 1 and n_altri >= 5:
        return altri * 0.20
    return 0.0
```

## classi distinte per cani e gatti ma con una superclasse

```python
class Pet:
    def __init__(self, specie):
        self._specie  = specie
        self._sesso   = None

class Gatto(Pet):
    def __init__(self, razza, sesso):
        super().__init__('Gatto') 
        self._razza = razza
        self._sesso = sesso
    def __repr__(self):    
        return f"\nAnimale domestico di specie: {self._specie}, razza: {self._razza}, sesso: {self._sesso}"
        
class Cane(Pet):
    def __init__(self, razza, sesso):
        super().__init__('Cane') 
        self._razza = razza
        self._sesso = sesso
    def __repr__(self):
        return f"\nAnimale domestico di specie: {self._specie}, razza: {self._razza}, sesso: {self._sesso}"
        
t = Cane('Terrier', 'M')        
c = Cane('Carlino', 'F')        
g = Gatto('Soriano', 'F')    
```
