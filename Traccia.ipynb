```python
class Allievo:
    def __init__(self, nome, cognome):
        self.nome = nome
        self.cognome = cognome 
    def __repr__(self):
        return f"\nNome: {self.nome}, Cognome: {self.cognome}"
```

```python
monica = Allievo('Monica', 'Fiocchi')
```

```python
type(monica)
```

```python
nicola = Allievo('Nicola', 'Verdini')
```
